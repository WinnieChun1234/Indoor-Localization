```python
import pandas as pd
import numpy as np
```

```python
env1 = pd.read_pickle("data/data_env1.pickle")
env2 = pd.read_pickle("data/data_env2.pickle")
```

```python
env1_rssi = env1["rssi"]
env2_rssi = env2["rssi"]
env1_ap = env1["ap"]
env2_ap = env2["ap"]
env1_labels = env1["labels"]
env2_labels = env2["labels"]
print(env1_rssi.shape)
print(env2_rssi.shape)
print(env1_ap.shape)
print(env2_ap.shape)
print(env1_labels.shape)
print(env2_labels.shape)
```

```python
# for ap, take mean along the axis 1
env1_ap = env1_ap.mean(axis=1)  # (6, 2)
env2_ap = env2_ap.mean(axis=1)  # (6, 2)
env1_ap = np.tile(env1_ap, (env1_rssi.shape[0], 1, 1))
env2_ap = np.tile(env2_ap, (env2_rssi.shape[0], 1, 1))
print(env1_ap.shape)
print(env2_ap.shape)
```

```python
env1_df = pd.DataFrame()
# feature shape (n_samples, n_features)
# lable shape (n_samples, 2)
env1_df["AP_1_RSSI"] = env1_rssi[:, 0]
env1_df["AP_2_RSSI"] = env1_rssi[:, 1]
env1_df["AP_3_RSSI"] = env1_rssi[:, 2]
env1_df["AP_4_RSSI"] = env1_rssi[:, 3]
env1_df["AP_5_RSSI"] = env1_rssi[:, 4]
env1_df["AP_6_RSSI"] = env1_rssi[:, 5]
env1_df["AP_1_LOC_X"] = env1_ap[:, 0, 0]
env1_df["AP_1_LOC_Y"] = env1_ap[:, 0, 1]
env1_df["AP_2_LOC_X"] = env1_ap[:, 0, 0]
env1_df["AP_2_LOC_Y"] = env1_ap[:, 0, 1]
env1_df["AP_3_LOC_X"] = env1_ap[:, 0, 0]
env1_df["AP_3_LOC_Y"] = env1_ap[:, 0, 1]
env1_df["AP_4_LOC_X"] = env1_ap[:, 0, 0]
env1_df["AP_4_LOC_Y"] = env1_ap[:, 0, 1]
env1_df["AP_5_LOC_X"] = env1_ap[:, 0, 0]
env1_df["AP_5_LOC_Y"] = env1_ap[:, 0, 1]
env1_df["AP_6_LOC_X"] = env1_ap[:, 0, 0]
env1_df["AP_6_LOC_Y"] = env1_ap[:, 0, 1]

env1_df["X"] = env1_labels[:, 0]
env1_df["Y"] = env1_labels[:, 1]
print(env1_df.shape)
print(env1_df)
```

```python
env2_df = pd.DataFrame()
# feature shape (n_samples, n_features)
# lable shape (n_samples, 2)
env2_df["AP_1_RSSI"] = env2_rssi[:, 0]
env2_df["AP_2_RSSI"] = env2_rssi[:, 1]
env2_df["AP_3_RSSI"] = env2_rssi[:, 2]
env2_df["AP_4_RSSI"] = env2_rssi[:, 3]
env2_df["AP_5_RSSI"] = env2_rssi[:, 4]
env2_df["AP_6_RSSI"] = env2_rssi[:, 5]
env2_df["AP_1_LOC_X"] = env2_ap[:, 0, 0]
env2_df["AP_1_LOC_Y"] = env2_ap[:, 0, 1]
env2_df["AP_2_LOC_X"] = env2_ap[:, 0, 0]
env2_df["AP_2_LOC_Y"] = env2_ap[:, 0, 1]
env2_df["AP_3_LOC_X"] = env2_ap[:, 0, 0]
env2_df["AP_3_LOC_Y"] = env2_ap[:, 0, 1]
env2_df["AP_4_LOC_X"] = env2_ap[:, 0, 0]
env2_df["AP_4_LOC_Y"] = env2_ap[:, 0, 1]
env2_df["AP_5_LOC_X"] = env2_ap[:, 0, 0]
env2_df["AP_5_LOC_Y"] = env2_ap[:, 0, 1]
env2_df["AP_6_LOC_X"] = env2_ap[:, 0, 0]
env2_df["AP_6_LOC_Y"] = env2_ap[:, 0, 1]

env2_df["X"] = env2_labels[:, 0]
env2_df["Y"] = env2_labels[:, 1]
print(env2_df.shape)
print(env2_df)
```

```python
env1_df.to_csv("data/env1.csv", index=False)
env2_df.to_csv("data/env2.csv", index=False)
```
